# 📓 실습 02. Streamlit 웹 정비 포털 시뮬레이션 및 app.py 빌드

이 실습에서는 사용자 질의에 응답하여 CMP 설비의 텍스트 조치 가이드 및 관련 엔티티 네트워크 지도를 시각화하는 Streamlit 웹 소스코드를 파일로 추출하고 노트북 상에서 렌더링 과정을 테스트합니다.

```python
import os
import networkx as nx
import matplotlib.pyplot as plt

# 한글 폰트 적용
plt.rcParams['font.family'] = 'Malgun Gothic'

# 1. Streamlit 애플리케이션 파일 자동 작성
app_code = """
import streamlit as st
import networkx as nx
import matplotlib.pyplot as plt

st.title("🛠️ 반도체 CMP 설비 인지형 정비 검색 포털")
query = st.text_input("장애 센서 경보 혹은 고장 증상을 입력하세요:", "노즐 막힘")
if st.button("SOP 검색"):
    st.subheader("추천 조치 가이드")
    st.info("[SOP-01] 초순수로 노즐을 1.5 bar 압력으로 역방향 플러싱하십시오.")
"""

with open("app.py", "w", encoding="utf-8") as f:
    f.write(app_code.strip())
print("app.py 웹 어플리케이션 파일 생성 완료")
```

```python
# 2. 웹 시각화 렌더링 물리 엔진 테스트
G = nx.Graph()
G.add_edge("슬러리 노즐 막힘", "[SOP-01] 노즐 세척")
G.add_edge("[SOP-01] 노즐 세척", "초순수 (DI Water)")

fig, ax = plt.subplots(figsize=(6, 4))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='orange', node_size=1500, font_size=10, ax=ax)
plt.title("정비 연동 컨텍스트 렌더링 검증", fontsize=12)
plt.show()
```
